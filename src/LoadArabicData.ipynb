```python
import numpy as np
from PIL import Image
import zipfile
import io
from google.colab import drive
from sklearn.model_selection import train_test_split
drive.mount('/content/drive')
import os
import cv2
from sklearn.preprocessing import OneHotEncoder
```

```python
zip_path = '/content/drive/MyDrive/CS539: Handwritten Letter Language Classifier/data/raw/ArabicData.zip'
extraction_path = '/content/arabic_data'

# Unzip the file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extraction_path)

# List the contents of the extracted folder to verify
os.listdir(extraction_path)
```

```python
alphabet_path = os.path.join(extraction_path, 'alphabet')
os.listdir(alphabet_path)
```

```python
def process_images_labels(alphabet_path):
    data = []
    target = []

    for letter in os.listdir(alphabet_path):
        letter_path = os.path.join(alphabet_path, letter)

        if os.path.isdir(letter_path):
            for img_file in os.listdir(letter_path):


                img_path = os.path.join(letter_path, img_file)
                img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

                img_resized = cv2.resize(img, (28, 28))

                img_flattened = img_resized.flatten()

                data.append(img_flattened)
                target.append(letter)
    return np.array(data), np.array(target)

# Process the images and labels
data_matrix, target_matrix = process_images_labels(alphabet_path)
```

```python
X = data_matrix.reshape(data_matrix.shape[0], 28, 28, 1)

# one hot encoding map for languages
language_map = {
    'English': [1, 0, 0, 0],
    'Greek': [0, 1, 0, 0],
    'Russian': [0, 0, 1, 0],
    'Arabic': [0, 0, 0, 1],
}

y = np.array([language_map['Arabic'] for _ in range(X.shape[0])])
print(X.shape)
print(y.shape)
np.savez('/content/drive/MyDrive/CS539: Handwritten Letter Language Classifier/data/processed/', X=X, y=y)
```

```python
encoder = OneHotEncoder(sparse_output=False)
X_chars = np.array(data_matrix)
X_chars = X.reshape(X.shape[0], 28, 28, 1)
y_chars = np.array(target_matrix)
y_chars = np.array(target_matrix).reshape(-1, 1)
y_chars = encoder.fit_transform(y_chars)
print(X_chars.shape)
print(y_chars.shape)
np.savez('/content/drive/MyDrive/CS539: Handwritten Letter Language Classifier/data/processed/', X=X_chars, y=y_chars)
```
