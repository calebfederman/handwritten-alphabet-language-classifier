```python
import numpy as np
from PIL import Image
import zipfile
import io
from google.colab import drive
from sklearn.model_selection import train_test_split
drive.mount('/content/drive')
import os
from sklearn.preprocessing import OneHotEncoder
```

```python
zip_path = '/content/drive/MyDrive/English.zip'
unzipped_data = {}

with zipfile.ZipFile(zip_path, 'r') as z:
    for file_name in z.namelist():
        with z.open(file_name) as file:
            unzipped_data[file_name] = file.read()

# Loading Images
image_data = []
labels = []
label_map = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

for file_name, data in unzipped_data.items():
    folder_name = file_name.split('/')[0]

    if folder_name not in label_map:
        continue

    image = Image.open(io.BytesIO(data))

    # convert to 28 x 28 greyscale and normalize
    image = image.convert('L').resize((28, 28))
    image = np.array(image) / 255.0

    image_data.append(np.array(image))
    labels.append(folder_name)

unique_labels = np.unique(np.array(labels))
print("Unique labels:", unique_labels)

len(image_data)
```

```python
# Creating data matrices with correct sizes for cnn
X = np.array(image_data)
X = X.reshape(X.shape[0], 28, 28, 1)


# one hot encoding map for languages
language_map = {
    'English': [1, 0, 0, 0],
    'Greek': [0, 1, 0, 0],
    'Russian': [0, 0, 1, 0],
    'Arabic': [0, 0, 0, 1],
}

# All the data is English rn
y = np.array([language_map['English'] for _ in range(len(image_data))])
print(X.shape)
print(y.shape)
np.savez('/content/drive/MyDrive/CS539: Handwritten Letter Language Classifier/data/processed/', X=X, y=y)
```

```python
encoder = OneHotEncoder(sparse_output=False)
X_chars = np.array(image_data)
X_chars = X.reshape(X.shape[0], 28, 28, 1)
y_chars = np.array(labels)
y_chars = np.array(labels).reshape(-1, 1)
y_chars = encoder.fit_transform(y_chars)
print(X_chars.shape)
print(y_chars.shape)
np.savez('/content/drive/MyDrive/CS539: Handwritten Letter Language Classifier/data/processed/', X=X_chars, y=y_chars)
```
